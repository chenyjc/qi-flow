```python
import qlib
from qlib.data.dataset import TSDatasetH
from qlib.data.dataset.handler import DataHandlerLP
import pandas as pd
import numpy as np

# 初始化 qlib
qlib.init(provider_uri="~/.qlib/qlib_data/cn_data", region="cn")

# 配置 handler
data_handler_config = {
    "start_time": "2016-01-01",
    "end_time": "2025-12-30",
    "fit_start_time": "2016-01-01",
    "fit_end_time": "2025-09-30",
    "instruments": "csi300",
    "label": ["Ref($close, -2) / Ref($close, -1) - 1"],
}

# 创建 handler
from qlib.contrib.data.handler import Alpha158
handler = Alpha158(**data_handler_config)

# 获取数据
df = handler.fetch(col_set=["feature", "label"])

# 检查数据质量
print("数据形状:", df.shape)
print("\n特征数据 NaN 数量:", df["feature"].isna().sum().sum())
print("标签数据 NaN 数量:", df["label"].isna().sum())
print("\n特征数据 Inf 数量:", np.isinf(df["feature"].values).sum())
print("标签数据 Inf 数量:", np.isinf(df["label"].values).sum())

# 查看统计信息
print("\n特征数据统计:")
print(df["feature"].describe())

print("\n标签数据统计:")
print(df["label"].describe())
```
